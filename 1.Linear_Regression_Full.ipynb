```python
import pandas as pd   #importing pandas because it help for various operation in tabular form , also helps to see the meaningful insights in the table - i created the alias name for pandas as =>pd for ease to use multiple time 
data=pd.read_csv("Salary_Data.csv")       #here the dataset has been readed with the help of pandas library and stored it in the 'data' variable and also modification in this dataset does not affect the original dataaset
```

```python
data #here the data variable is printed for checking whether out imported dataset is correct or not
```

```python
# Input and Output split :
feature = data[["YearsExperience"]]      # here the feature or input column is taken from the dataset-'data' and stored it in 'feature' variable
target = data[["Salary"]]     # here the output or target column is taken from the dataset-'data' and stored it in 'target' variable

# the input and output data has been splitted successfully
```

```python
feature  # here just checking the input is correctly assigned in a 'feature'variable to reduce further errors 
```

```python
target       # here just checking the output is correctly assigned in a 'target'variable to reduce further errors 
```

```python
# spliting of training and testing dataset 

from sklearn.model_selection import train_test_split       # from scikit learn library - inside model_selection class, we have imported the 'train_test_split' for seperating or spitting the training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(feature,target,test_size=.20,random_state=0)      #here the 'train_test_split'-method occupies four parameters , 1.input , 2.output , 3.size of the test dataset to be split(test_size) , 4.random_state

#It returns the four different datasets , so we have created four different variables to store those datasets 

#the splitting of training and testing dataset has been successfully executed and splitted ....
```

```python
x_train.count()  # just printing how many rows the 'x_train' variable consist of and what is the datatype present inside it 
```

```python
x_test.count()       # just printing how many rows the 'x_test' variable consist of and what is the datatype present inside it 
```

```python
y_train.count()          # just printing how many rows the 'y_train' variable consist of and what is the datatype present inside it 
```

```python
y_test.count()          # just printing how many rows the 'y_test' variable consist of and what is the datatype present inside it 
```

```python
#                                "  _______MODEL CREATION PHASE ________  "

# after splitting training and testing datasets , we need to create a model with the help of trainning dataset ...
# here we are going to create a model with training dataset


from sklearn.linear_model import LinearRegression   # from 'scikitlearn'-library , inside linear_model class , we imported 'LinearRegression'- it is an algorithm used to predict the variables given by end users using training dataset 
regressor=LinearRegression()    # just storing the 'LinearRegression'-(algorithm or class) in the variable called 'regressor'
regressor   #just printing the variable to see weather the method is stored in a assigned variable or not
```

```python
# creation of model using training dataset => also called as training of the model

model_crt=regressor.fit(x_train,y_train)     #inside 'linearregression'- algorithm, there is a (method or function) called 'fit' which helps to train our model using training dataset ==>> we have already assigned the 'LinearRegression' in a variable so , using such variable with dot operator(.) to call the function  and store the trained model in another variable
model_crt     # just checking weather the trained model is stored in same variable or not  


# Model has been created successfully inside the variable called 'model_crt'....
```

```python
# SLR (simple linear regression) formula :" y = w x + b "
# after creating the model => that is , filling out all the variables except input data on above formula , just checking weather the weight'w' and bias'b' is calculated or not 
weight = model_crt.coef_       #using coef_ method to see the variable of weight'w'
bias = model_crt.intercept_    #using intercept_ method to see the variable of weight'w'
```

```python
weight    #just printing what is present inside the weight'w' variable
```

```python
bias     #just printing what is present inside the bias'b' variable
```

```python
#               """"   ___________DEPLOYMENT PHASE__________ """"

# testing stage : 
#after creating the model using training dataset , we have to test the model using the testing inputs => here 'x_test'

y_pred=model_crt.predict(x_test)    # inside 'linearregression' class , there is a method called 'predict' - with the help of predict method , your model will get tested with 'x_test'-input for any kind of prediction and stored it in another variable

# our AI model gets generated .....
```

```python
y_pred        # just printing the output of the predicted model using 'x_test'-input
```

```python
# after testing the model ,we have to know the value of predicted output  is correct or not , so we use r2_score to calculate the accuracy of our tested model- 'predicted output'

from sklearn.metrics import r2_score   # using scikit learn - metrics class => importing r2_score  ->  the r2_score will ranges from 0 to 1 : if the accuracy or result is near to 0 implies our model is not learned well , if its near to 1 then good model -> proceed to end users 
r_score = r2_score(y_test,y_pred)      # the r2_score method consist of two parameters to calculate the accuracy of predicted output from actual output (i.e)=>(actual_output-(y_test) and predicted_output-(variable of the predicted output - here(y_pred)))
```

```python
r_score   # just seeing how accuracy is our model is ... It ranges from 0 to 1 : if the accuracy or result is near to 0 implies our model is not learned well , if its near to 1 then good model -> proceed to endusers
```

```python
# SAVE THE BEST MODEL 
# the accuracy of our model is near to 1 , so we have to save this model and give to front end developer and made usable for end user

import pickle as pk    # pickle library is used for saving the model'created' inside python ide
store=pk.dump(model_crt,open("finalyzed_model.sav","wb"))     #we have to save out created model for further use -> dump method in pickle library is used for dumping-up or inserting  the model created in python 
# we have successfully saved the model in the name of : "finalyzed_model.sav"  , and "wb"-writebinary is nothing but creation of model used wb to write inside it 
```

```python
load=pk.load(open("finalyzed_model.sav","rb"))  # just loaded the saved model for just checking up => deployment checkup 
```
